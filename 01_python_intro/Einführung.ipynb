### if-Anweisung

```python
i = input('Geben Sie einen einstelligen positiven Wert ein:  ')    # datatype str
i = int(i)

if (i < 5) and (i>= 0):
    print('Bin kleiner als fünf!  - Wert: ', i)

elif i == 5:
    print('Ich bin fünf! - Wert: ', i)

elif (i > 5) and (i < 10):
    print('Ich bin größer als fünf! - Wert: ', i)
    
else:
    print('Du hast keinen einstelligen positiven Wert eingegeben! - Wert: ',i)
   
```

```python
# combine with 5 loops:

for j in range(5):
    
    i = input('Geben Sie einen einstelligen positiven Wert ein:  ')    # datatype str
    i = int(i)

    if (i < 5) and (i>= 0):
        print('Bin kleiner als fünf!  - Wert: ', i)

    elif i == 5:
        print('Ich bin fünf! - Wert: ', i)

    elif (i > 5) and (i < 10):
        print('Ich bin größer als fünf! - Wert: ', i)

    else:
        print('Du hast keinen einstelligen positiven Wert eingegeben! - Wert: ',i)
   
```

```python
# to do a break after wrong input

for j in range(5):
    
    i = input('Geben Sie einen einstelligen positiven Wert ein:  ')    # datatype str
    i = int(i)

    if (i < 5) and (i>= 0):
        print('Bin kleiner als fünf!  - Wert: ', i)

    elif i == 5:
        print('Ich bin fünf! - Wert: ', i)

    elif (i > 5) and (i < 10):
        print('Ich bin größer als fünf! - Wert: ', i)

    else:
        print('Du hast keinen einstelligen positiven Wert eingegeben! - Wert: ',i)
        print('Deine Versuche werden abgebrochen!')
        break
```
