## Trigonometrische Funktionen

Hierzu geh√∂ren die Sinus, Cosinus und Tangens Funktionen.

```python
import matplotlib.pyplot as plt

def trigonomy_plot(m, n, trig):
    x1 = np.linspace(0, 2*np.pi, m)
    x2 = np.linspace(0, 2*np.pi, n)
    x = [x1, x2]
    
    fig, ax = plt.subplots(1, 2, figsize = (10, 3))
    
    for i, ax in enumerate(ax.flat):
        if trig == 'sin':
            y = np.sin(x[i])
        if trig == 'cos':
            y = np.cos(x[i])
        if trig == 'tan':
            y = np.tan(x[i])
        if trig == 'arcsin':
            y = np.arcsin(x[i])
        if trig == 'arccos':
            y = np.arccos(x[i])
        if trig == 'arctan':
            y = np.arctan(x[i])
        ax.scatter(x[i], y)
        ax.set_title(trig)
        plt.xlim(0.0, 2*np.pi)
        plt.ylim(-1.0, 1.0);
    
```

```python
# sinus
trigonomy_plot(6, 100, 'sin')
```

```python
# cosinus
trigonomy_plot(6, 100, 'cos')
```

```python
# tangens
trigonomy_plot(6, 100, 'tan')
```

```python
def trigonomy_line(m):
    x = np.linspace(0, 4*np.pi, m)
    trig = ['sin', 'cos']
    
    plt.figure(figsize = (20, 3))
    
    for i in range(len(trig)):
        if trig[i] == 'sin':
            y = np.sin(x)
        if trig[i] == 'cos':
            y = np.cos(x)
        plt.plot(x, y)
        #plt.set_title(trig[i])
        plt.xlim(0.0, 4*np.pi)
        plt.ylim(-1.0, 1.0);
```

```python
trigonomy_line(100)
```
