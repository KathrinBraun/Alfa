## Statistische Maßzahlen: Lageparameter



### 6. mean()

Gibt den arithmetischen Mittelwert wieder:

$$ \bar{x} = \frac{1}{n} \sum_{i=0}^n x_i = \frac{x_0 + x_1 + ... + x_n}{n}$$


```python
np.mean(x_int)
```

```python
np.mean(x_norm)
```

### 7. median()

Für den Median sortiert man die Daten in eine Rangliste $x_1^*, x_2^*, ...., x_n^*$ an. Der Median ist dann der Wert, der in der Mittel liegt (bei gerade Anzahl wird das mittel zwischen den beiden mittig liegenden Werte genommen):

$$x_{med} = \left\{ \begin{array}{ll} \frac{x_\frac{n}{2}^* + x_{\frac{n}{2}+1}^*}{2} ~, falls \ n \ gerade \\ 
                     x_{\frac{n+1}{2}}^* ~, falls \ n \ ungerade \end{array} \right.$$



```python
np.median(x_int)
```

```python
np.median(x_norm)
```

### 8. mode()

Der Modalwert (auch Modus genannt) gibt die Anzahl des häufigst genannten Wert an.

```python
# creating out of x_rand an array only with integer number
x_norm_int = x_norm.round()
x_norm_int[0:10]
```

## Statistische Maßzahlen: Streuungsparameter



### 1. var()

Bei der Varianz wird im Prinzip die mittlere quadratische Abweichung der Beobachtungswerte vom arithmetischen Mittel bestimmt.

Bei der **empirischen Varianz** wird bei der Division durch n-1 bei der Anzahl bereinigt. Diese wird bei reinen Zufallsstichproben verwendet:

$$ s^2 = \frac{1}{n-1}\sum_{i=1}^n \left( x_i - \bar{x} \right) $$

In der Regel wird die Grundgesamtheit als Beobachtungsreihe betrachtet und daher wird die **Grundgesamtheitsvarianz** bestimmt. Diese wird mit $s_0^2$ bezeichnet. Bei dieser Varianz wird auch von **Maximum Likelihood** gesprochen.

$$ s_0^2 = \frac{1}{n}\sum_{i=1}^n \left( x_i - \bar{x} \right) $$

Die empirische Varianz ist i.d.R. erwartungsgetreuer als die Grundgesamtheitsvarianz.

Auf NumPy.org ist bei der Methode var() folgende Information zu finden:



Das bedeutet, dass der Standard $s_0^2$ berechnet und durch ddof = 1 zu einer empirischen Varianz verändert werden kann.


```python
# Maximum Likelihood
np.var(x_norm)
```

```python
# empirical Variance
np.var(x_norm, ddof = 1)
```

### 2. std()

Die Standardabweisung $s$ ist die Wurzel der Varianz

```python
# standard deviation
np.std(x_norm)
```

```python
# take root of variance
np.var(x_norm)**(1/2)
```

```python
# take root of varian by using np.sqrt
np.sqrt(np.var(x_norm))
```

### 3. Spannweite: max()-min()

Die Spannweite beschreibt den Abstand zwischen dem größten und kleinsten Beobachtungswert.

$$ SP = max(x) - min(x) $$

```python
np.max(x_norm) - np.min(x_norm)
```

### 4. percentile()

Hier wird der Quatilsabstand zwischen den beiden empirischen Quartilen bestimmt.

Bei dieser Methode muss noch das q eingegeben werden, wo das Quantil bestimmt werden soll. Dieser Wert muss zwischen 0 und 100 liegen. Bei q = 50 ist dies gleich dem Median.


```python
# q = 50
np.percentile(x_norm, q = 50)
```

```python
np.median(x_norm)
```

```python
# 2 sided quantile
np.percentile(x_norm, q = (25, 75))
```

**Summary of Aggregations**

    np.sum() / np.nansum() - Summe der Elemente berechnen.
    np.prod() / np.nanprod() - Produkt der Elemente berechnen.
    np.max() / np.nanmax() - Maximum bestimmen.
    np.min() / np.nanmin() - Minimum finden.
    np.mean() / np.nanmean() - Mittelwert berechnen.
    np.std() / np.nanstd() - Standardabweichung berechnen.
    np.var() / np.nanvar() - Varianz berechnen.
    np.median() / np.nanmedian() - Median bestimmen.
    np.percentile() / np.percentile() - Bestimmt das Quantil.
    np.argmin() / np.nanargmin() - Bestimmt den Index mit dem Minimalwert.
    np.argmax() / np.nanargmax() - Bestimmt den Index mit dem Maximalwert.
    np.any() / n.v. - Testen, ob irgendein Element wahr ist.
    np.all() / n.v. - Testen, ob alle Elemente wahr sind.
