## Refresh: Vektoren



    Bildquelle: learnattack.de
    
Vektoren kann man sich als Pfeile vorstellen: sie haben eine gewisse Länge und eine Richtung in der geschossen wird. Ein dreidimensionaler Vektor wird folgendermaßen dargestellt:

$$\vec{a} = (a_1, a_2, a_3) =  \left(\begin{array}{c} a_1 \\ a_2 \\ a_3 \end{array}\right) $$

#### 1. Addition

$$\vec{a} + \vec{b} = \left(\begin{array}{c} a_1 \\ a_2 \\ a_3 \end{array}\right) + \left(\begin{array}{c} b_1 \\ b_2 \\ b_3 \end{array}\right)= \left(\begin{array}{c} a_1 + b_1 \\ a_2 + b_2 \\ a_3 + b_3 \end{array}\right)$$

```python
# define vector a and b
a = np.array([2, 3, 5])
b = np.array([7, 4, 2])
```

```python
# addition
a + b
```

#### 2. Subtraktion

$$\vec{a} - \vec{b} = \left(\begin{array}{c} a_1 \\ a_2 \\ a_3 \end{array}\right) - \left(\begin{array}{c} b_1 \\ b_2 \\ b_3 \end{array}\right)= \left(\begin{array}{c} a_1 - b_1 \\ a_2 - b_2 \\ a_3 - b_3 \end{array}\right)$$

```python
# subtraction
a - b
```

#### 3. Betrag eines Vektors

Die Länge eines Vektors vom Ende bis zur Pfeilspitze wird Betrag genannt. Dieser kann durch den **Satz des Pythagoras** ermittelt werden:

$$c^2 = a^2 + b^2 \Longrightarrow c = \sqrt{a+b} $$ 



    Bildquelle: www.br.de


$$a = \left|\vec{a} \right|  = \sqrt{a_{1}^{2} + a_{2}^{2} + a_{3}^{2}}  $$

Ist der Betrag $a = 0$ so wird dieser Vektor auch **Nullvektor** $\vec{0}$ bezeichnet.

Ein Vektor mit dem Betrag $a = 1$ wird **Einheitsvektor** genannt.

```python
# using the method sqrt from NumPy 
np.sqrt(a[0]**2 + a[1]**2 + a[2]**2)
```

#### 4. Multiplikation eines Vektors mit einer reellen Zahl

$$ c \cdot \vec{a} = c \cdot \left(\begin{array}{c} a_1 \\ a_2 \\ a_3 \end{array}\right) = \left(\begin{array}{c} c \cdot a_1 \\ c \cdot a_2 \\ c \cdot a_3 \end{array}\right) $$

   * c > 1: Vektor wird gestreckt
   * 0 < c < 1: Vektor wird gestaucht
   * c < 0: Vektor wird gesteckt bzw. gestaucht und die Pfeilspitze zeigt in die andere Richtung
    
    

```python
# multiplication of real number with vector
4 * a
```

#### 5. Skalarprodukt

$$ \vec{a} \cdot \vec{b} = \left(\begin{array}{c} a_1 \\ a_2 \\ a_3 \end{array}\right) \cdot
                           \left(\begin{array}{c} b_1 \\ b_2 \\ b_3 \end{array}\right) =
                           a_1 \cdot b_1 + a_2 \cdot b_2 + a_3 \cdot b_3 $$
                           

Definiert ist das Skalarprodut durch die Multiplikation ihrer Beträge mit dem Cosinus des eingeschlossenen Winkels:

$$ \vec{a} \cdot \vec{b} = \left|\vec{a}\right| \cdot \left|\vec{b}\right| \cdot \cos \alpha $$

```python
# scalar product (or dot product)
a * b
```

```python
# to get the end result, we have to add the values together
sum(a*b)
```

```python
# using function dot()
np.dot(a, b)
```

#### 6. Kreuzprodukt

    

    Bildquelle: wikipedia.org
    


    Bildquelle: mathenexus.zum.de
                                                   
<br>
<br>

$$\vec{a} \times \vec{b} = \left(\begin{array}{c} a_1 \\ a_2 \\ a_3 \end{array}\right) \times
                           \left(\begin{array}{c} b_1 \\ b_2 \\ b_3 \end{array}\right) =
                            \left(\begin{array}{c} a_2 b_3 - a_3 b_2 \\ 
                                                   a_3 b_1 - a_1 b_3 \\
                                                   a_1 b_2 - a_2 b_1 \end{array}\right)$$

                                                   


```python
# cross product
np.cross(a, b)
```

```python
# determine area |axb| area
cross_vec = np.cross(a,b)
np.sqrt(cross_vec[0]**2 + cross_vec[1]**2 + cross_vec[2]**2)
```

### 8. Determinante (Spatprodukt)

Gegeben seien 2 Vektoren $a_1$ und $a_2$: 
$$ \vec{a_1} = \left( \begin{array}{r} 1 \\ 2 \\ \end{array} \right) ~ und ~
\vec{a_2} = \left( \begin{array}{r} 3 \\ 4 \\ \end{array} \right)  $$ 

Hiervon wollen wir nun die Determinante bestimmen:

$$det(a_1, a_2) = \left| \begin{array}{rr} 1 & 3 \\ 2 & 4 \\ \end{array} \right| $$

Hieraus ergibt sich folgende Rechnung:

$$det(a_1, a_2) = 1 \cdot 4 - 3 \cdot 2 = -2 $$

```python
# putting both vectors together as matrix
A = np.array([[1, 3], [2, 4]])
A
```

```python
# determine det by using np.linalg.det(A)
det = np.linalg.det(A)
det
```

## Trigonometrische Funktionen

Hierzu gehören die Sinus, Cosinus und Tangens Funktionen. 

```python
import matplotlib.pylab as plt

def trigonomy_plot(m, n, trig):
    x1 = np.linspace(0, 2*np.pi, m)
    x2 = np.linspace(0, 2*np.pi, n)
    x = [x1, x2]
    
    fig, ax = plt.subplots(1, 2, figsize = (10, 3))
    
    for i, ax in enumerate(ax.flat):
        if trig == 'sin':
            y = np.sin(x[i])
        if trig == 'cos':
            y = np.cos(x[i])
        if trig == 'tan':
            y = np.tan(x[i])
        if trig == 'arcsin':
            y = np.arcsin(x[i])
        if trig == 'arccos':
            y = np.arccos(x[i])
        if trig == 'arctan':
            y = np.arctan(x[i])
        ax.scatter(x[i], y)
        ax.set_title(trig)
        plt.xlim(0.0, 2*np.pi)
        plt.ylim(-1.0, 1.0);
    
```

```python
# sinus
trigonomy_plot(6, 100, 'sin')
```

```python
# cosinus
trigonomy_plot(6, 100, 'cos')
```

```python
# tangens
trigonomy_plot(6, 100, 'tan')
```

```python
def trigonomy_line(m):
    x = np.linspace(0, 4*np.pi, m)
    trig = ['sin', 'cos']
    
    plt.figure(figsize = (20, 3))
    
    for i in range(len(trig)):
        if trig[i] == 'sin':
            y = np.sin(x)
        if trig[i] == 'cos':
            y = np.cos(x)
        plt.plot(x, y)
        #plt.set_title(trig[i])
        plt.xlim(0.0, 4*np.pi)
        plt.ylim(-1.0, 1.0);
```

```python
trigonomy_line(100)
```
