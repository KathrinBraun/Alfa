## Matplotlib module pyplot

### Linestyle

```python
# back to our 3 lines previously
x = np.linspace(0,1,100)
for i in [1,2,3]:
    y = x ** i
    plt.plot(x,y, label = f"power = {i}");
    
plt.legend();
```

```python
# add linestyle in
for i in [1,2,3]:
    y = x ** i
    plt.plot(x,y, label = f"power = {i}", linestyle = '--');
    
plt.legend();
```

```python
# each line should have their own style and color is gray
# create tupels for i and style
for i, style in [(1, ':'), (2, '--'), (3, '-.')]:
    y = x ** i
    plt.plot(x,y, label = f"power = {i}", linestyle = style, color = 'gray');
    
plt.legend();
```

```python
# play with linewidth
for i, style in [(1, ':'), (2, '--'), (3, '-.')]:
    y = x ** i
    plt.plot(x,y, label = f"power = {i}", linestyle = style, color = 'gray', linewidth = 4);
    
plt.legend();
```

### Meta

```python
# add title, xlabel, ylabel...

for i, style in [(1, ':'), (2, '--'), (3, '-.')]:
    y = x ** i
    plt.plot(x,y, label = f"power = {i}", linestyle = style, color = 'gray', linewidth = 4);
    
    plt.title('The three gray lines!')
    plt.xlabel('x values')
    plt.ylabel('y values')
    
plt.legend();
```

```python
# shape of figure is the set first!
plt.figure(figsize = (7,7))

for i, style in [(1, ':'), (2, '--'), (3, '-.')]:
    y = x ** i
    plt.plot(x,y, label = f"power = {i}", linestyle = style, color = 'gray', linewidth = 4);
    
    plt.title('The three gray lines!')
    plt.xlabel('x values')
    plt.ylabel('y values')
    
plt.legend();
```

### Subplot without ending of 's'

```python
# several plots in on plot

plt.figure(figsize = (15, 8))

for i, mue in enumerate([0, 1, 2]):
    for j, sigma in enumerate([0.1, 0.2, 0.3]):
        
        x = np.random.normal(mue, sigma, (1000))
        plt.hist(x)
```

```python
# each plot in a separate plot by using subplot
```

```python
plt.figure(figsize = (15, 8))

for i, mue in enumerate([0, 1, 2]):
    for j, sigma in enumerate([0.1, 0.2, 0.3]):
        
        x = np.random.normal(mue, sigma, (1000))
        
        plt.subplot(3, 3, 3*i+j+1)
        plt.hist(x)
```

```python
# to smooth xlim, title with using 'Latex'

plt.figure(figsize = (15, 8))

for i, mue in enumerate([0, 1, 2]):
    for j, sigma in enumerate([0.1, 0.2, 0.3]):
        
        x = np.random.normal(mue, sigma, (1000))
        
        plt.subplot(3, 3, 3*i+j+1)
        plt.hist(x)
        plt.xlim(-1, 3)
        plt.title(f"$\mu$ = {mue}, $\sigma$ = {sigma}")
```

### Subplots with 's'

```python
# using subplots instead of figure 
fig, ax = plt.subplots(3, 3,figsize = (15, 8))

for i, mue in enumerate([0, 1, 2]):
    for j, sigma in enumerate([0.1, 0.2, 0.3]):
        
        x = np.random.normal(mue, sigma, (1000,))
        
        ax[i,j].hist(x, 30)
        ax[i,j].set_title(f"$\mu$ = {mue}, $\sigma$ = {sigma}")  # change to set_title
        ax[i,j].set_xlim(-1, 3)                                  # change to set_xlim      
```

```python
# share x and y
fig, ax = plt.subplots(3, 3,figsize = (15, 8), sharex = True, sharey = True)

for i, mue in enumerate([0, 1, 2]):
    for j, sigma in enumerate([0.1, 0.2, 0.3]):
        
        x = np.random.normal(mue, sigma, (1000,))
        
        ax[i,j].hist(x, 30)
        ax[i,j].set_title(f"$\mu$ = {mue}, $\sigma$ = {sigma}")  # change to set_title
        ax[i,j].set_xlim(-1, 3)                                  # change to set_xlim      
```

### Function

```python
def grid_plots(mues, sigmas):
    
    # change grid with len of input
    fig, ax = plt.subplots(len(mues), len(sigmas),figsize = (15, 8), sharex = True, sharey = True)
                             
    # for loop is running now over the input
    for i, mue in enumerate(mues):
        for j, sigma in enumerate(sigmas):

            x = np.random.normal(mue, sigma, (1000,))

            ax[i,j].hist(x, 30)
            ax[i,j].set_title(f"$\mu$ = {mue}, $\sigma$ = {sigma}") 
            ax[i,j].set_xlim(-1, 3)                                     
```

```python
grid_plots([-1, 0, 1, 3], [0.1, 0.3, 0.4])
```

```python
# adjust xlim and colors
def grid_plots(mues, sigmas):
    
    color = ['DarkTurquoise', 'Teal', 'MediumBlue']
    
    fig, ax = plt.subplots(len(mues), len(sigmas),figsize = (15, 8), sharex = True, sharey = True)
                             
    for i, mue in enumerate(mues):
        for j, sigma in enumerate(sigmas):

            x = np.random.normal(mue, sigma, (1000,))

            ax[i,j].hist(x, 30, color = color[j])
            ax[i,j].set_title(f"$\mu$ = {mue}, $\sigma$ = {sigma}") 
            ax[i,j].set_xlim(min(mues) - max(sigmas)*3,
                             max(mues) + max(sigmas)*3)     
```

```python
grid_plots([-1, 0, 1, 3], [0.1, 0.3, 0.4])
```
