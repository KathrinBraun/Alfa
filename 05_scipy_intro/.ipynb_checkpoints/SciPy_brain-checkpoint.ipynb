### SciPy: Analysis of female and male brain

```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# load data
data = pd.read_csv('brain_size.csv', delimiter = ';')
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
# del Unnamed: 0
data = data.drop('Unnamed: 0', axis = 1)
```

```python
data.head()
```

```python
data.info()
```

```python
# Weight and Height are object because we have some '.' in
# determine how many '.' are in the data
data_dot_weight = data[(data.Weight == '.') == True]
data_dot_weight
```

```python
# we decide to delete both person
data = data[(data.Weight == '.') == False]
data.info()
```

```python
# reject of '.' in Heigth
data_dot_height = data[(data.Height == '.') == True]
data_dot_height
```

```python
#transorm object into float
data['Height'] = data['Height'].astype(float)
data['Weight'] = data['Weight'].astype(float)
data.info()
```

```python
sns.pairplot(data=data, hue='Gender', palette = ['blue', 'green'], hue_order = ['Male', 'Female']);
```

```python
data.columns
```

```python
sns.heatmap(data[(data.Gender == 'Male') == True][['FSIQ', 'VIQ', 'PIQ', 'Weight', 'Height', 'MRI_Count']].corr() , 
            cmap = 'Blues');
```

```python
sns.heatmap(data[(data.Gender == 'Female') == True][['FSIQ', 'VIQ', 'PIQ', 'Weight', 'Height', 'MRI_Count']].corr() , 
            cmap = 'Greens');
```

```python
# 1-sample Test
def one_sample_test(data, feature, gender):
    
    data_gender = data[(data.Gender == gender) == True]
    
    for i in feature:
        res = stats.ttest_1samp(data_gender[i], 0)
        print(i, ' - ', res)


```

```python
feature = ['FSIQ', 'VIQ', 'PIQ', 'Weight', 'Height', 'MRI_Count']
one_sample_test(data, feature, 'Male')
```

```python
one_sample_test(data, feature, 'Female')
```
