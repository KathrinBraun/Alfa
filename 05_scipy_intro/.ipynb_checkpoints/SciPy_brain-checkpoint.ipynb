```python
# 1-sample Test
def one_sample_test(data, feature):
    '''one_sample_test
    
    Input: data (df), feature (list)
    
    The 1-sample Test is testing the mean of a sample against a mean of another sample. Therefor we will test a Male feature against a 
    the same Female feature.'''
    
    # split in dataframe regarding Male and Female
    data_male = data[(data.Gender == 'Male') == True]
    data_female = data[(data.Gender == 'Female') == True]
    
    
    for j, i in enumerate(feature):
        # ttest_1samp(data, popmean)
        res = stats.ttest_1samp(data_male[i], data_female[i].mean())  # Second input is here a mean
        #plt.subplot(2, 3, j+1)
        #plt.hist(data[i], 10)
        print(i, ' - t-Statistic: ', res[0], '   p-Value: ', res[1])
        
    return
```

```python
feature = ['FSIQ', 'VIQ', 'PIQ', 'Weight', 'Height', 'MRI_Count']
one_sample_test(data, feature)
```

Lediglich FSIQ, VIQ und PIQ haben einen größeren p-Wert als 0.05 - unterscheiden sich also nicht. 

Bei dem Gewicht und der Größe unterscheiden sich jedoch die Mittelwerte voneinander.

```python
# 2-sample Test
def two_sample_test(data, feature):
    '''two_sample_test
    
    Input: data (df), feature (list)
    
    The 2-sample Test is testing the signifikanz of a sample against a mean of another sample. 
    This is a two-sided test for the null hypothesis that 2 independent samples
    have identical average (expected) values. This test assumes that the
    populations have identical variances by default.
    '''
    
    # split in dataframe regarding Male and Female
    data_male = data[(data.Gender == 'Male') == True]
    data_female = data[(data.Gender == 'Female') == True]
    
    
    for j, i in enumerate(feature):
        # ttest_ind(data, data)
        res = stats.ttest_ind(data_male[i], data_female[i])
        print(i, ' - t-Statistic: ', res[0], '   p-Value: ', res[1])
        
    return
    
```

```python
two_sample_test(data, feature)
```

Der 2-Stichproben-Test zeigt in etwa dem Ergebnis des 1-Stichproben-Tests: FSIQ, VIQ und PIQ (p-Wert > 0.5) unterscheiden sich nicht.

#### Ergebnis

Bei FSIQ, VIQ und PIQ ist der p-Wert > 0.05, d.h. die Differenz zwischen den Mittelwerten ist statistisch nicht signifikant und wir verwerfen nicht $H_0$.

In der Größe und im Gewicht ist jedoch der p-Wert < 0.05 und daher ist die Differenz zwischen den Mittelwerten signifikant. Hier verwerfen wir $H_0$ ebenso wie bei dem MRI_Count.
