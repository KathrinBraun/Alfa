## Take Me Home Challenge: Iris

Der Blumenhändler "Iris blüht" möchte die Iris Sorten (Schwertlilien) Setosa, Versicolor und Virginica vertreiben. Leider hat er derzeit wenig Kenntnis über die Blumen. Er hat jedoch sich Bilder ausgeben lassen:

| **Setosa** | **Versicolor** | **Virginica** |
| --- | --- | ---|
|![title](iris_setosa.jpg) | ![title](iris_versicolor.jpg)| ![title](iris_virginica.jpg)


Er erkennt nur, dass alle Iris Sorten lila sind. Nun hat er von einem Konkurrenten freundlicherweise eine Liste erhalten, über seine Verkauften Iris Sorten letztes Jahr im Mai. Mit dieser Liste kommt er nun zu Ihnen.

<br>
<br>
Die Liste enthält folgende Informationen:
<br>

| Name | Beschreibung |
|---|---|
| Sepal Length | Länge der Kelchblätter |
| Sepal Width | Breite der Kelchblätter |
| Pedal Length | Länge der Blütenblätter |
| Pedal Width | Breite der Blütenblätter |
| Ort | einstellige Ziffern |
| Anzahl | Verkaufte Stückzahl |

<br>
<br>
Nach längerem Bitten erhält er auch noch die Informationen, welche Ortschaften sich hinter den Ziffern befinden:
<br>

| Zahl | Ort |
| --- | --- |
| 0 | München |
| 1 | Rostock |
| 2 | Trier |
| 3 | Münster |
| 4 | Ulm |
| 5 | Hamburg |
| 6 | Köln |
| 7 | Würzburg |
| 8 | Leipzig |
| 9 | Wolfsburg |

<br>
<br>
Desweiteren möchte der Blumenhändler seinen Verkaufspreis anhand der Länge der Blütenblätter festlegen, da er meint, dass dies das entscheidende Qualitätskriterium ist.
<br>

| Petal Width in CM | Preis pro Stück |
| --- | --- |
| PW < 1 | 2.55 |
| 1 <= PW < 2 | 2.66 |
| 2 <= PW < 3 | 2.68 |
| 3 <= PW < 4 | 2.72 |
| 4 <= PW < 5 | 2.77 |
| 5 <= PW < 6 | 2.85 |
| 6 <= PW < 7 | 2.90 |
| 7 <= PW  | 3.00 |

<br>
<br>

Da der Blümenhändler eifrig sein möchte, möchte er auch die Planzen selber züchten. Er rechnet mit folgenden Kosten:

| Sorte | Kosten pro Stück |
| --- | --- |
| Setosa | 1.03 |
| Versicolor | 0.98 |
| Virginica | 1.55 |




### Aufgabe

1. Ersetzen Sie die Zahlen mit den jeweiligen Orte.
2. Der Blumenhändler hat schon festgestellt, dass ein paar Informationen über die Kelch- und / oder Blütenblätter fehlen. Er möchte nicht, dass diese Elemente gelöscht werden, sondern durch den jeweiligen Median ersetzt werden. (Hinweis: Ausreißer sind ansonsten hier nicht zu erwarten.)
3. Informieren Sie den Blumenhändler, wie sich die Irise unterscheiden.
4. Visualisieren Sie, ob die Merkmale der Kelch- und Blütenblätter korrelieren.
5. Bestimmen Sie rechnerisch, wie auch visuell, welche Sorte, die Häufigkeit und an welchem Ort verkauft wurde.

#### Zusatz 1
6. Ermitteln Sie den Preis, Kosten und Gewinn je Sorte.
7. Ermitteln Sie den Gewinn an den jeweiligen Standorten pro Sorte.
8. Können Sie eine Empfehlung für die Zucht und Verkauf den Blumenhändler anbieten?
