Der Winzer Vinho Verde hat eine Qualitätskontrolle durchführen lassen. Die Kontrolleure haben eine Datei für den Rotwein und eine Datei für den Weißwein erstellt. Die Beschreibung der Daten können Sie der Textdatei winequality.names entnehmen.

Zudem stellt er eine Liste bereit, über die Produktion der Weine von 1991 - 1998 für die jeweiligen Counties. Der Winzer hat uns mitgeteilt, dass die Weine den gleichen Namen haben, wie die Counties. In jedem County sind jeweils weiße und rote Trauben für die jeweiligen Weine angebaut worden. In der Liste sind die produzierten Liter pro County enthalten.

Der Winzer möchte eine Übersicht erhalten, welche Weine (County) gut und weniger gut von den Experten eingestuft wurden.

Mindestens folgende Fragen möchte er beantwortet haben:

1. Wie viele Weine (rot/weiß) sind in den jeweiligen Qualitätsklassen?
2. Welche Qualitätsklasse hat pro Q-Merkmal den größten bzw. den kleinsten Median. Sehen Sie dies als positiv oder negativ?
3. Wieviel Liter (rot/weiß) sind im Jahre "1998" in den jeweiligen Qualtitätsklassen produziert worden?
4. Welche Weine (Counties) beurteilen die Experten als gut bzw. schlecht ein?

Zusatz:

5. Bestimmen Sie die 3 besten und 3 schlechtesten Weine (Counties) anhand der Qualitätsklassen. Wie ist die Produktionsentwicklung seit 1991 bis inkl. 1998?

## 1. Wie viele Weine (rot/weiß) sind in den jeweiligen Qualitätsklassen?

```python
fig, ax = plt.subplots(1,2, figsize = (15, 5))
df_white.groupby('quality').count()['alcohol'].plot(kind = 'bar', ax = ax[0], color = 'orange');
df_red.groupby('quality').count()['alcohol'].plot(kind = 'bar', ax = ax[1], color = 'darkred');
```

## 2. Welches Qualitätsklasse hat pro Q-Merkmal den größten bzw. den kleinsten Median. Sehen Sie dies als positiv oder negativ?

```python
df_white.groupby(['quality']).median()[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
```

Ergebnis zu 2.: **Whitewine**

| Q-Merkmal | Max | Min| Pos / Neg |
| --- | --- | --- | --- |
| fixed acidity | 7 | 1 | Medians differ only 0.4 - neutral |
| volatile acidity | 6+7 | 4 | 0.32 in Q4 seams not good |
| citric acid | 9 | 4 | Higher value seams positiv |
| residual sugar | 5 | 9 | Smaller value seams positiv |
| clorides | 6 | 9 | Smaller value seams positiv |
| free sulfur dioxide | 8+5 | 4 | The values changes not in a pattern |
| total sulfur dioxide | 3 | 4 | The max tends to be negativ. see next boxplot |
| density | 8 | 9 | Medians differ only 0.003 - neutral |
| ph | 9 | 4+5 | Medians differ only 0,12 neutral |
| sulphates | 6+7 | 3 | Medians differ only 0,04 |
| alcohol | 9 | 5 | Higher median seams to be positiv |

## 3. Wieviel Liter (rot/weiß) sind im Jahre "1998" in den jeweiligen Qualtitätsklassen produziert worden?

```python
df_white.columns
```

```python
temp = ['BRANCO_91', 'BRANCO_92', 'BRANCO_93', 'BRANCO_94','BRANCO_95', 'BRANCO_96', 'BRANCO_97', 'BRANCO_98',]

for i in temp:
    df_white[i] = df_white[i].astype(int)
```

```python
print('In 1998 Vinho Verde produced:', df_white.BRANCO_98.sum(), 'liter')
```

```python
df_red.columns
```

```python
temp = ['TINTO_91', 'TINTO_92', 'TINTO_93', 'TINTO_94', 'TINTO_95', 'TINTO_96', 'TINTO_97', 'TINTO_98']

for i in temp:
    df_red[i] = df_red[i].astype(int)
```

```python
print('In 1998 Vinho Verde produced:', df_red.TINTO_98.sum(), 'liter')
```
