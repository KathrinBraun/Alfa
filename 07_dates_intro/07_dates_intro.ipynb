In Anlehnung an Vanderplas Kap. 3.12

Varianten von Kalenderdaten und Zeitreihen:

* **Zeitstempel**: verweisen auf einen bestimmten Zeitpunkt. (3. März 2021 um 13:45 Uhr)
* **Zeitintervall**: bezeichnen eine Zeitspannen zwischen einem Anfangs- und Endzeitpunkt liegt. (30 Stunden)
* **Zeitdelta / -dauer**: kennzeichnen exakte Zeitabschnitte.(vom 01.02.2000 - 31.03.2018)

### Zeitreihen in Pandas

In Pandas können die Zeitstempel als Index benutzt werden. Somit wird bei einem Merkmal (z.B. Kurs) in einer Series von Pandas gearbeitet.

```python
# index by using DatetimeIndex
index = pd.DatetimeIndex(['22-03-2020', '23-03-2020', '24-03-2020', '25-03-2020'])
data = pd.Series([0, 1, 2, 3], index = index)
data
```

```python
# as well possible with '.'
index = pd.DatetimeIndex(['22.03.2020', '23.03.2020', '24.03.2020', '25.03.2020'])
data = pd.Series([0, 1, 2, 3], index = index)
data
```

```python
# slicing by using index
data['2020-03-23': '2020-03-25']
```

```python
data['2020']
```

### Datenstrukturen für Zeitreihen in Pandas

Datentypen in Pandas:

* **Timestamp für Zeitstempel**: dieser beruht auf den Datentyp numpy.datetime64. Die dazugehörige Indexstruktur heißt **DatetimeIndex**.
* **Period für Zeitperioden**: speichert ein numpy.datetime64-Zeitintervall festgelegter Länge. Die dazugehörige Indexstruktur heißt **PeriodIndex**.
* **Timedelta für Zeitdeltas oder -dauer**: beruht ebenfalls auf numpy.datetime64.

```python
# to_datetime()
dates = pd.to_datetime([datetime(2020, 7, 3), '3rd of January 2020', '2020-Aug-3', '08-08-2020', '20200615' ])
dates
```

```python
# to_period()
dates.to_period('M')
```

```python
dates.to_period('D')
```

```python
# subtraction
dates - dates[0]
```

#### Resampling und Anpassung der Häufigkeiten

**Resampling**: Zeitreihen mit erhöhter oder verringerter Häufigkeit neu einzulesen. Hierbei kann *resample()* (Datenaggregation) oder auch die einfachere Methode *asfreq()* (Datenauswahl) verwendet werden.

Bei beiden Methoden ist voreingestellt, dass NA-Werte bei NA-Werte bleiben. Zum Füllen wird *fillna()* benötigt

```python
goog.plot(alpha = 0.3, style = '-', color = 'blue', figsize = (10, 8))
goog.resample('BA').mean().plot(style= ':', color = 'purple', label = 'Mean last year', linewidth = 6)            # BA End of Business Year
goog.asfreq('BA').plot(style = '--', color = 'darkturquoise', label = 'Value beginning of the year', linewidth = 6 )
plt.legend()
```

```python
fig, ax = plt.subplots(1, 2, sharex = True, figsize = (20, 5))
data = goog.iloc[:15]
data.asfreq('D').plot(ax=ax[0], marker = 'o')
data.asfreq('D', method = 'bfill').plot(ax=ax[1], style = '-o', color = 'darkblue')   # filling backwards: method = 'bfill'
data.asfreq('D', method ='ffill').plot(ax=ax[1], style = '-o', color = 'darkgreen')   # filling forwards: method = 'ffil'
ax[1].legend(["filling backwards", "filling forwards"]);
```

### Zeitliches Verschieben

**shift()** verschiebt die Daten

**tshift()** verschiebt den Index

```python
fig, ax = plt.subplots(3, sharey=True, figsize=(15, 10))

# assign data frequency
goog = goog.asfreq('D', method='pad')

# plot 
goog.plot(ax=ax[0])
goog.shift(1000).plot(ax=ax[1])   # shifting of data
goog.tshift(1000).plot(ax=ax[2])  # shifting of index

# legends and notes
local_max = pd.to_datetime('2012-06-25')
offset = pd.Timedelta(900, 'D')
ax[0].legend(['input'], loc=2)
ax[0].get_xticklabels()[4].set(weight='heavy', color='red')
ax[0].axvline(local_max, alpha=0.3, color='red')
ax[1].legend(['shift(1000)'], loc=2)
ax[1].get_xticklabels()[4].set(weight='heavy', color='red')
ax[1].axvline(local_max + offset, alpha=0.3, color='red')
ax[2].legend(['tshift(1000)'], loc=2)
ax[2].get_xticklabels()[1].set(weight='heavy', color='red')
ax[2].axvline(local_max + offset, alpha=0.3, color='red');
```

```python
# shifting for calculation of ROI (Kapitalmarktrendite)
ROI = 100 * (goog.tshift(-365) / goog - 1)
ROI.plot(figsize = (20, 5))
plt.ylabel('ROI in %');
```

### Glättung von Zeitreihen

**rolling()** liefert eine ähnliche Ansicht wie die **groupby()**-Operation.

```python
rolling = goog.rolling(365, center=True)
rolling
```

```python
data = pd.DataFrame({'input': goog,'Annual Mean Value': rolling.mean(), 'Standard Deviation': rolling.std()})
data.head()
```

```python
# annual mean value and standard deviation
rolling = goog.rolling(365, center=True)
data = pd.DataFrame({'input': goog,'Annual Mean Value': rolling.mean(), 'Standard Deviation': rolling.std()})
ax = data.plot(style=['-', '--', ':'], color = 'darkgreen', figsize = (15, 8), linewidth = 4)
ax.lines[0].set_alpha(0.3)
```
